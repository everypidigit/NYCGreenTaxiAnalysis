## Let's work with sklearn now
We will import necessary functions and will split data

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
colsWithMissingVals = []

for column in workDF.columns:
    have = workDF[column].isnull().values.any()
    if have == True:
        colsWithMissingVals.append(column)
colsWithMissingVals
```

```python
target = 'trip_distance'

y = workDF[target]
X = workDF.drop(target, axis=1).copy()
```

```python
X_train_full, X_test_full, y_train, y_test = train_test_split(X, y, 
                                                                train_size=0.8, test_size=0.2,
                                                                random_state=0)
```

### We remember that there are some missing values in columns
We also know that some columns contain categorical data, while others are numerical ones. 
I will imput missing values depending on a given column type 

```python
# finding the categorical columns
categorical_cols = []
for column in X_train_full.columns:
    if X_train_full[column].dtype == "object":
        categorical_cols.append(column)
        
# finding the numerical columns
numerical_cols = []
for column in X_train_full.columns:
    if X_train_full[column].dtype in ['int64', 'float64']:
        numerical_cols.append(column)
print(categorical_cols, numerical_cols)
```

```python
# at first, let's try to train our models only with numerical and categorical columns

cols = categorical_cols + numerical_cols
X_train = X_train_full[cols].copy()
X_test = X_test_full[cols].copy()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_absolute_error
```

```python
numerical_transformer = SimpleImputer(strategy='most_frequent')

# Preprocessing for categorical data
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
# let's define our models
# I will define 4 different models and will try all of them to compare results

RFRModel = RandomForestRegressor(n_estimators=25, random_state=0)
XGBoostModel = XGBRegressor(n_estimators = 50, random_state=0)
HistGBRModel = HistGradientBoostingRegressor(learning_rate = 0.05, random_state=0)
GBRModel = GradientBoostingRegressor(n_estimators = 10, learning_rate = 0.05, random_state=0)
```

```python
models = [RFRModel, XGBoostModel, HistGBRModel, GBRModel]

results = {}

for model in models:
    pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                          ('model', model)
                         ])

    pipeline.fit(X_train, y_train)
    preds = pipeline.predict(X_test)
    
    mae = mean_absolute_error(y_test, preds)

    results[model] = mae
    print(f'for {model} MAE is: {mae}', )
```

```python
# plotting a bar chart to visualize different models' results
# although it is not very pretty...

plt.figure(figsize=[27,12])
plt.bar(range(len(results)), list(results.values()))
plt.xticks(range(len(results)), list(results.keys()))
```

## Now, let's try to do cross-validation in order to see if we can improve results
Or, possibly, find some issues with existing ones...

```python
from sklearn.model_selection import cross_val_score
```

```python
crossValResults = {}

for model in models:
    crossvalPipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model)])
    
    scores = -1 * cross_val_score(crossvalPipeline, X, y,
                              cv=5,
                              scoring='neg_mean_absolute_error')
    mae = scores.mean()
    crossValResults[model] = mae

    print(f"For {model} average MAE score: {mae}")
```

```python
crossValResults
```

```python
plt.figure(figsize=[27,12])
plt.bar(range(len(crossValResults)), list(crossValResults.values()))
plt.xticks(range(len(crossValResults)), list(crossValResults.keys()))
```

### It seems like the random forest models has the best results:
at least withouth hyperparameter tuning. these results are also more robust since we used crossvalidation

## Now, let's try to use shuffle split for cross validation and see if there is something interesting in that

```python
from sklearn.model_selection import ShuffleSplit
```

```python
shuffleCrossValResults = {}
cv = ShuffleSplit(n_splits=8, test_size=0.25, random_state=0)

for model in models:
    crossvalPipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model)])
    
    scores = -1 * cross_val_score(crossvalPipeline, X, y, cv=cv,
                              scoring='neg_mean_absolute_error')
    mae = scores.mean()
    shuffleCrossValResults[model] = mae

    print(f"For {model} average MAE score: {mae}")
```

```python
shuffleCrossValResults
```

```python
plt.figure(figsize=[30,12])
plt.bar(range(len(shuffleCrossValResults)), list(shuffleCrossValResults.values()))
plt.xticks(range(len(shuffleCrossValResults)), list(shuffleCrossValResults.keys()))
```

### Seems like shuffling data when using crossvalidation did not help at all:
the predictions are much worse compared to usual crossvalidation. it can be attributed to the possibility
of models learning some dependepcies on some data and then testing them on batches on data that has slightly different
dependencies. it has to be due to data shuffling
